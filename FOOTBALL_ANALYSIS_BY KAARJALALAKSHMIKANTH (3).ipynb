# Exploratory Data Analysis (EDA)

- EDA is an important part of any project involving data.
- It is important to investigate and understand the data better before building a model with it.
- A few questions have been mentioned below which will help you approach the analysis in the right manner and generate insights from the data.
- A thorough analysis of the data, in addition to the questions mentioned below, should be done.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel("C:/Users/Lakshmi kanth/Desktop/foot_ball/football_data_KL.xlsx")  # after merage dataset
```

# 1.What is the distribution of home and away goals?

```python
sns.histplot(df['home_club_goals'], kde=True, color='blue', label='Home Goals')
sns.histplot(df['away_club_goals'], kde=True, color='red', label='Away Goals')
plt.legend()
plt.show()
```

**conclustion:-**
* Overall, they have scored around 1,200 goals at home, compared to around 800 away goals. 
* This suggests that the team performs better when playing at home.

# 2.Which stadium has the highest average attendance?

```python
avg_attendance = df.groupby('stadium')['attendance'].mean().sort_values(ascending=False)
print(avg_attendance.head(5))
```

```python
df1= pd.DataFrame(avg_attendance) # i am connverting data frame
df1  
```

**conclustion:-**
* Old Trafford stadium is highest attendance there.
* MSV-Arena stadium is very less attendance there.

```python
top_yellow_cards =df.groupby('player_name')['yellow_cards'].sum().sort_values(ascending=False)
print(top_yellow_cards.head(5))
```

```python
top_yellow_card= pd.DataFrame(top_yellow_cards.head())
```

```python
top_yellow_card
```

# 3.Who are the top 5 players with the most yellow cards?

```python
top_yellow_cards = df.groupby('player_name')['yellow_cards'].sum().sort_values(ascending=False)
print(top_yellow_cards.head(5))
```

**conclustion:-**
* John Anthony Brooks is highest yellow cards.    

# 4.Is there any relationship between home club position and the number of goals scored?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='home_club_position', y='home_club_goals', color='purple')
plt.xlabel('Home Club Position')
plt.ylabel('Home Club Goals')
plt.show()
```

**conclustion:-**
* The scatter plot shows a positive correlation between home club position and home club goals.    

# 5.What is the distribution of minutes played by players?

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='minutes_played', bins=20, kde=True, color='green')
plt.xlabel('Minutes Played')
plt.ylabel('Frequency')
plt.show()
```

**conclustion:-**
* Most people play for less than 20 minutes, and the number of people who play steadily decreases as the play time increases.
* There are very few people who play for more than 80 minutes.    

# 6.How many unique referees and managers are there?

```python
unique_referees = df['referee'].nunique()
unique_managers = pd.concat([df['home_club_manager_name'], df['away_club_manager_name']]).nunique()
print(f"Unique Referees: {unique_referees}, Unique Managers: {unique_managers}")
```

# 7.What is the distribution of goals scored by home and away teams?

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='home_club_goals', bins=20, color='blue', label='Home Team Goals')
sns.histplot(data=df, x='away_club_goals', bins=20, color='red', label='Away Team Goals')
plt.xlabel('Goals')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**conclustion:-**
* This is in line with the concept of home advantage in sports, which suggests that teams have an advantage when playing at home.

# 8.What is the correlation between red cards and goals?

```python
correlation = df['red_cards'].corr(df['goals'])
print(correlation)
```

# 9.Which manager has the highest win rate for home games?

```python
home_wins = df[df['home_club_goals'] > df['away_club_goals']].groupby('home_club_manager_name').size()
home_games = df.groupby('home_club_manager_name').size()
win_rate = (home_wins / home_games).sort_values(ascending=False)
print(win_rate.head(5))
```
